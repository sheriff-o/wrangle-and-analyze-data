# Insights and Visualizations from Wrangling the WeRateDogs data


### INTRODUCTION

The WeRateDogs Enhanced Twitter archive contains data from 2356 of the 5000+ tweets from the @dog_rates twitter account, posted between November 15, 2015 and August 1, 2017. The data consist of dog ratings which were taken the text of the tweets along with the dog name and dog stage present.

The retweet_count and favourite_count for each tweet were not included in the enhanced archive, hence the need to download the additional data from twitter account separately.

Also separately downloaded, was the image prediction filefrom Udacity's servers which contains the top three predictions for dog breeds based on the images from the tweets.


### Wrangling Data

The data gathered from the three sources were assessed visually and programmatically for quality and tidiness issues. Some of the issues were documented and then cleaned programmatically using the recommended approach - Define, Code and Test after a copy has been made. Thereafter, the files were merged to form a single file saved with the name "twitter_archive_master.csv".

Some few analysis was carried out on this final data.


### INSIGHTS

1. There were total of three breed predictions. For the first prediction, there are 1333 dogs and 165 non-dogs, 1343 dogs and      155 non-dogs for the second prediction, the 1498 dogs and 0 non-dogs for the third prediction. 

2. The dog with the highest rating is Logan, with a tweet_id of '786709082849828864'

3. The dog which has the highest retweet_count has no name as it is None.

```python
# Relationships between rating and favorite_count
twitter_archive_master.plot(x='rating', y='favorite_count', kind='scatter', title='Ratings vs Favourite Count');
```
